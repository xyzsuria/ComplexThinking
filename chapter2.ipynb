```python
import networkx as nx
G = nx.DiGraph()
```

```python
G.add_node('Alice')
G.add_node('Bob')
G.add_node('Chuck')
list(G.nodes())
```

```python
G.add_edge('Alice', 'Bob')
G.add_edge('Alice', 'Chuck')
G.add_edge('Bob', 'Alice')
G.add_edge('Bob', 'Chuck')
print(G.edges())
```

```python
nx.draw_circular(G,
                  node_color='red',
                  node_size=2000,
                  with_labels=True)
```

```python
positions = dict(Albany=(-74, 43),
                  Boston=(-71, 42),
                  NYC=(-74, 41),
                  Philly=(-75, 40))
G = nx.Graph()
G.add_nodes_from(positions)
drive_times = {('Albany', 'Boston'): 3,
                ('Albany', 'NYC'): 4,
                ('Boston', 'NYC'): 4,
                ('NYC', 'Philly'): 2}
G.add_edges_from(drive_times)
nx.draw(G, positions,
        node_color='blue',
        node_shape='s',
        node_size=2500,
        with_labels=True)
```

```python
nx.draw_networkx_edge_labels(G, positions,edge_labels=drive_times)
```

```python
def all_pairs(nodes):
    for i, u in enumerate(nodes):
        for j, v in enumerate(nodes):
              if i>j:
                  yield u, v
```

```python
def make_complete_graph(n):
    G = nx.Graph()
    nodes = range(n)
    G.add_nodes_from(nodes)
    G.add_edges_from(all_pairs(nodes))
    return G
```

```python
complete = make_complete_graph(10)
nx.draw_circular(complete,
                  node_color='grey',
                  node_size=1000,
                  with_labels=True)
```
